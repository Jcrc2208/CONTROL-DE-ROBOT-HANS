```
import yaml

# Definir la configuración limpia para YOLO
data_yaml_content = {
    'path': '/content/CONTROL-DE-ROBOT-HANS/PiezaMetalica/data',  # Ruta base
    'train': 'images/train',                                      # Carpeta de entrenamiento
    'val': 'images/val',                                          # Carpeta de validación
    'names': {
        0: 'piezaMetalica'                                        # Tu clase única
    }
}

# Escribir el archivo data.yaml con el formato correcto en la raíz del proyecto
ruta_yaml = '/content/CONTROL-DE-ROBOT-HANS/data.yaml'
with open(ruta_yaml, 'w') as f:
    yaml.dump(data_yaml_content, f, default_flow_style=False, sort_keys=False)

print("¡Archivo data.yaml reescrito y corregido con éxito!")
```

```
from ultralytics import YOLO

# 1. Cargamos el archivo de respaldo que subiste (el cerebro del Epoch 24)
model = YOLO('/content/CONTROL-DE-ROBOT-HANS/last.pt')

# 2. Reanudamos el entrenamiento con 'resume=True'
# (No necesitas poner 'epochs' ni 'imgsz' porque YOLO los lee en automático desde el last.pt)
results = model.train(
    data='/content/CONTROL-DE-ROBOT-HANS/data.yaml',   # Tu archivo de configuración
    resume=True                                        # ¡La bandera de la salvación!
)
```
